```python
# Import data sets
import pandas as pd
import os

LEAGUE_PATH = "datasets/League"

def load_summer_data(league_path=LEAGUE_PATH):
    csv_path = os.path.join(league_path, "2018_Summer_League.csv")
    return pd.read_csv(csv_path, delimiter=',', encoding="utf-8-sig")
    
def load_spring_data(league_path=LEAGUE_PATH):
    csv_path = os.path.join(league_path, "2018_Spring_League.csv")
    return pd.read_csv(csv_path, delimiter=',', encoding="utf-8-sig")

def load_2018_worlds_data(league_path=LEAGUE_PATH):
    csv_path = os.path.join(league_path, "2018_worlds.csv")
    return pd.read_csv(csv_path, delimiter=',', encoding="utf-8-sig")
```

```python
# Pandas custom options.
pd.options.display.max_columns = 200
pd.options.display.max_rows = 200
```

```python
# Set data sets to variables.
league_summer = load_summer_data()
league_spring = load_spring_data()
league_worlds_2018 = load_2018_worlds_data()
league_worlds_2018.head()
```

```python
# Builds out sets per team/region
# NA
kaisa_c9 = kaisa_data.loc[kaisa_data['team'] == 'Cloud9']
kaisa_liquid = kaisa_data.loc[kaisa_data['team'] == 'Team Liquid']
kaisa_100 = kaisa_data.loc[kaisa_data['team'] == '100 Thieves']
na_kaisa = pd.concat([kaisa_c9, kaisa_liquid, kaisa_100], axis=0)

# LCK
kaisa_afreeca = kaisa_data.loc[kaisa_data['team'] == 'Afreeca Freecs']
kaisa_gen = kaisa_data.loc[kaisa_data['team'] == 'Gen.G']
kaisa_kt = kaisa_data.loc[kaisa_data['team'] == 'KT Rolster']
lck_kaisa = pd.concat([kaisa_afreeca, kaisa_gen, kaisa_kt], axis=0)

# LPL
kaisa_ig = kaisa_data.loc[kaisa_data['team'] == 'Invictus Gaming']
kaisa_edg = kaisa_data.loc[kaisa_data['team'] == 'Edward Gaming']
kaisa_rng = kaisa_data.loc[kaisa_data['team'] == 'Royal Never Give Up']
lpl_kaisa = pd.concat([kaisa_ig, kaisa_edg, kaisa_rng], axis=0)

# EU

kaisa_g2 = kaisa_data.loc[kaisa_data['team'] == 'G2 Esports']
kaisa_vit = kaisa_data.loc[kaisa_data['team'] == 'Vitality']
kaisa_fnc = kaisa_data.loc[kaisa_data['team'] == 'Fnatic']
eu_kaisa = pd.concat([kaisa_g2, kaisa_vit, kaisa_fnc], axis=0)
```

```python
# NA win rate
na_kaisa['result'].sum() / len(na_kaisa.index)
```

```python
# LPL Win Rate
lpl_kaisa['result'].sum() / len(lpl_kaisa.index)
```
