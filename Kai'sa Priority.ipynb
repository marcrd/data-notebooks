```python
# Import data sets
import pandas as pd
import os

LEAGUE_PATH = "datasets/League"

def load_summer_data(league_path=LEAGUE_PATH):
    csv_path = os.path.join(league_path, "2018_Summer_League.csv")
    return pd.read_csv(csv_path, delimiter=',', encoding="utf-8-sig")
    
def load_spring_data(league_path=LEAGUE_PATH):
    csv_path = os.path.join(league_path, "2018_Spring_League.csv")
    return pd.read_csv(csv_path, delimiter=',', encoding="utf-8-sig")

def load_2018_worlds_data(league_path=LEAGUE_PATH):
    csv_path = os.path.join(league_path, "2018_worlds.csv")
    return pd.read_csv(csv_path, delimiter=',', encoding="utf-8-sig")
```

```python
# Pandas custom options.
pd.options.display.max_columns = 200
pd.options.display.max_rows = 200
```

```python
# Set data sets to variables.
league_summer = load_summer_data()
league_spring = load_spring_data()
league_worlds_2018 = load_2018_worlds_data()
league_worlds_2018.head()
```

```python
# Build out games by team and region
# As a note, as of quarters, no regional team has played each other in the same match, so we can assume the aggregate
# games here are all unique games.
games_c9 = df.loc[(df['team'] == 'Cloud9') & (df['player'] == 'Team')]
games_liquid = df.loc[(df['team'] == 'Team Liquid') & (df['player'] == 'Team')]
games_100 = df.loc[(df['team'] == '100 Thieves') & (df['player'] == 'Team')]
na_games = pd.concat([games_c9, games_liquid, games_100], axis=0)

# LCK
games_afreeca = df.loc[(df['team'] == 'Afreeca Freecs') & (df['player'] == 'Team')]
games_gen = df.loc[(df['team'] == 'Gen.G') & (df['player'] == 'Team')]
games_kt = df.loc[(df['team'] == 'KT Rolster') & (df['player'] == 'Team')]
lck_games = pd.concat([games_afreeca, games_gen, games_kt], axis=0)

# LPL
games_ig = df.loc[(df['team'] == 'Invictus Gaming') & (df['player'] == 'Team')]
games_edg = df.loc[(df['team'] == 'Edward Gaming') & (df['player'] == 'Team')]
games_rng = df.loc[(df['team'] == 'Royal Never Give Up') & (df['player'] == 'Team')]
lpl_games = pd.concat([games_ig, games_edg, games_rng], axis=0)

# EU

games_g2 = df.loc[(df['team'] == 'G2 Esports') & (df['player'] == 'Team')]
games_vit = df.loc[(df['team'] == 'Vitality') & (df['player'] == 'Team')]
games_fnc = df.loc[(df['team'] == 'Fnatic') & (df['player'] == 'Team')]
eu_games = pd.concat([games_g2, games_vit, games_fnc], axis=0)
```
